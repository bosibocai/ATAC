scRNA-seq注释文件预处理
- 读取h5ad文件`adata-fixFemale.h5ad`
- 切分每个样本的注释文件
- 读取细胞类型`majorType-fix`，并保存csv

```python
filePath = "/data/xwt/adata-fixFemale.h5ad"
adata = anndata.read(filePath)
adata
```

```python
adata.obs
```

scRNA-seq数据没有18T

```python
batches = ['10N', '10T', '11T', '12N', '12T', '13T', '19T', '20T', '21T', '22T']
```

```python
savePath = "./data/"
```

```python
for batch in batches:
    adata_batch = adata[adata.obs['batch'] == batch]
    metadata = adata_batch.obs[['majorType-fix']].rename(columns={'majorType-fix': 'majorType'})
    metadata['cell'] = metadata.index
    metadata['cell'] = metadata['cell'].str.replace(f'-{batch}', '')
    metadata.to_csv(f'{savePath}{batch}.csv', index=False)
```
